```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Constants
initial_investment = 1000
num_periods = 100
num_trials = 10000
```

```python
# Possible returns
returns = [2, 0.25]

# Generate Monte Carlo simulation
final_values = []
for _ in range(num_trials):
    wealth = initial_investment
    for _ in range(num_periods):
        wealth *= np.random.choice(returns, p=[0.5, 0.5])
    final_values.append(wealth)
```

```python
# Plotting the histogram of the final wealth values
plt.figure(figsize=(10, 6))
plt.hist(final_values, bins=50, color='blue', alpha=0.7)
plt.xlabel('Final Wealth')
plt.ylabel('Frequency')
plt.title('Histogram of Final Wealth after 100 Periods')
plt.grid(True)
plt.show()
```

```python
# Calculating percentiles
percentiles = [1, 5, 10, 25, 50, 75, 90, 95, 99, 99.99, 99.999, 99.9999]
percentile_values = np.percentile(final_values, percentiles)

# Creating a table with percentile values
percentile_table = pd.DataFrame({
    'Percentile': percentiles,
    'Final Wealth': percentile_values
})

# Formatting the 'Final Wealth' column as dollars without decimals
percentile_table['Final Wealth'] = percentile_table['Final Wealth'].apply(lambda x: f'${x:,.2f}')

# Print the table
print(percentile_table)
```
